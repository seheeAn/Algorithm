##### 2.1 binary search

```python
def bsSearch(data,item,low,high):
    mid=(low+high)//2
    if (low <= high):
        if(data[mid]==item):
            return mid
        elif(data[mid]>item):
            high = mid-1
            return(bsSearch(data,item,low,high))
        else:
            low = mid+1
            return(bsSearch(data,item,low,high))
    else:
        return -1

data = [1,3,5,6,7,9,10,14,17,19]
n=10
bsSearch(data,17,0,n-1)
```

(class.ver)

```python
class data():
    def __init__(self,data):
        self.data=data
        self.low = 0
        self.high = len(data)-1
        
    def binsearch(self,key):
        mid=(self.low+self.high)//2
        if (self.low <= self.high):
            if(self.data[mid]==key):
                return mid
            elif(self.data[mid]>key):
                self.high = mid-1
                return(self.binsearch(key))
            else:
                self.low = mid+1
                return(self.binsearch(key))
        else:
            return -1

data1 = data([1,3,5,6,7,9,10,14,17,19])
data1.binsearch(10)
```

measure average search count

```python
import random 

def bsSearch(data,item,low,high):
    global N
    mid=(low+high)//2
    if (low <= high):
        N+=1
        if(data[mid]==item):
            return mid
        elif(data[mid]>item):
            high = mid-1
            return(bsSearch(data,item,low,high))
        else:
            low = mid+1
            return(bsSearch(data,item,low,high))
    else:
        return -1

data = [1,3,5,6,7,9,10,14,17,19]
n=10
Nlist=[0]*20

for i in range(20):
    N=0
    item = random.randrange(1,19)
    bsSearch(data,item,0,n-1)
    Nlist[i] = N

Nsum = sum(Nlist)
print(Nsum/20)
```

##### 2.2 mergesort

```python
def mergesort(n,s):
    h=n//2
    m=n-h
    if(n>1):
        u=s[:h]
        v=s[h:n]
        mergesort(h,u)
        mergesort(m,v)
        merge(h,m,u,v,s)
        
def merge(h,m,u,v,s):
    i=j=k=0
    while i<h and j<m:
        if u[i] < v[j]:
            s[k]=u[i]
            i+=1
        else:
            s[k]=v[j]
            j+=1
        k+=1
    if i>= h:
        for idx in range(m-j):
            s[k+idx] = v[j+idx]
    elif j>=m:
        for idx in range(h-i):
            s[k+idx]=u[i+idx]

s=[3,5,2,9,10,14,4,8]
mergesort(8,s)
print(s)
```

##### in-place sort.ver (공간복잡도가 개선된 merge함수)

```python
def mergesort2(s,low,high):
    if(low<high):
        mid=(low+high)//2
        mergesort2(s,low,mid)
        mergesort2(s,mid+1,high)
        merge2(s,low,mid,high)
        
def merge2(s,low,mid,high):
    i=low
    j=mid+1
    k=0
    u = [0]*(high-low+1)
    while i<=mid and j<=high:
        if s[i] < s[j]:
            u[k]= s[i]
            i+=1
        else:
            u[k]=s[j]
            j+=1
        k+=1
    if i>mid:
        u[k:]= s[j:]
    else:
        u[k:]=s[i:mid+1]
    s[low:]=u[:]

s=[3,5,2,9,10,14,4,8]
mergesort2(s,0,7)
print(s)
```

##### 2.3 QuickSort

```python
def quickSort(s,low,high):
    if(high>low):
        pivotpoint = partition(s,low,high)
        quickSort(s,pivotpoint+1,high)
        quickSort(s,low,pivotpoint-1)

def partition(s,low,high):
    pivot = s[low]
    j=low
    i = low+1
    for i in range(low+1,high+1):
        if pivot > s[i]:
            j+=1
            s[i],s[j] = s[j],s[i]   
    s[j], s[low] = s[low],s[j]
    return j
    
s=[1,10,7,4,5,6,2,14]
quickSort(s,0,7)
print(s)
```

##### 2.2 mergesort

```python
def mergesort(n,s):
    h=n//2
    m=n-h
    if(n>1):
        u=s[:h]
        v=s[h:n]
        mergesort(h,u)
        mergesort(m,v)
        merge(h,m,u,v,s)
        
def merge(h,m,u,v,s):
    i=j=k=0
    while i<h and j<m:
        if u[i] < v[j]:
            s[k]=u[i]
            i+=1
        else:
            s[k]=v[j]
            j+=1
        k+=1
    if i>= h:
        for idx in range(m-j):
            s[k+idx] = v[j+idx]
    elif j>=m:
        for idx in range(h-i):
            s[k+idx]=u[i+idx]

s=[3,5,2,9,10,14,4,8]
mergesort(8,s)
print(s)
```

##### in-place sort.ver (공간복잡도가 개선된 merge함수)

```python
def mergesort2(s,low,high):
    if(low<high):
        mid=(low+high)//2
        mergesort2(s,low,mid)
        mergesort2(s,mid+1,high)
        merge2(s,low,mid,high)
        
def merge2(s,low,mid,high):
    i=low
    j=mid+1
    k=0
    u = [0]*(high-low+1)
    while i<=mid and j<=high:
        if s[i] < s[j]:
            u[k]= s[i]
            i+=1
        else:
            u[k]=s[j]
            j+=1
        k+=1
    if i>mid:
        u[k:]= s[j:]
    else:
        u[k:]=s[i:mid+1]
    s[low:]=u[:]

s=[3,5,2,9,10,14,4,8]
mergesort2(s,0,7)
print(s)
```

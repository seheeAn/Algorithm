##### 2.3 QuickSort

```python
def quickSort(s,low,high):
    if(high>low):
        pivotpoint = partition(s,low,high)
        quickSort(s,pivotpoint+1,high)
        quickSort(s,low,pivotpoint-1)

def partition(s,low,high):
    pivot = s[low]
    j=low
    i = low+1
    for i in range(low+1,high+1):
        if pivot > s[i]:
            j+=1
            s[i],s[j] = s[j],s[i]   
    s[j], s[low] = s[low],s[j]
    return j
    
s=[1,10,7,4,5,6,2,14]
quickSort(s,0,7)
print(s)
```

평균 시간복잡도 계산

```python
import random
    
def quickSort(s,low,high):
    if(high>low):
        pivotpoint = partition(s,low,high)
        quickSort(s,pivotpoint+1,high)
        quickSort(s,low,pivotpoint-1)
    
def partition(s,low,high):
    global N
    pivot = s[low]
    j=low
    i = low+1
    for i in range(low+1,high+1):
        N+=1
        if pivot > s[i]:
            j+=1
            s[i],s[j] = s[j],s[i]
    s[j], s[low] = s[low],s[j]
    return j
    
Nlist = [0]*5
    
for k in range(0,5):
    Nsum = 0
    n=8*(k+1)
    s=[0]*n #n=8,16,24,32,40
    for i in range(20): #20 times
        N=0
        for j in range(n):
            s[j] = random.randint(1,n+1)
        quickSort(s,0,n-1)
        Nsum += N
    Nlist[k] = Nsum/20
    
print(Nlist)
```

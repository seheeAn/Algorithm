##### 2.1 binary search

```python
def bsSearch(data,item,low,high):
    mid=(low+high)//2
    if (low <= high):
        if(data[mid]==item):
            return mid
        elif(data[mid]>item):
            high = mid-1
            return(bsSearch(data,item,low,high))
        else:
            low = mid+1
            return(bsSearch(data,item,low,high))
    else:
        return -1

data = [1,3,5,6,7,9,10,14,17,19]
n=10
bsSearch(data,17,0,n-1)
```

```python
class data():
    def __init__(self,data):
        self.data=data
        self.low = 0
        self.high = len(data)-1
        
    def binsearch(self,key):
        mid=(self.low+self.high)//2
        if (self.low <= self.high):
            if(self.data[mid]==key):
                return mid
            elif(self.data[mid]>key):
                self.high = mid-1
                return(self.binsearch(key))
            else:
                self.low = mid+1
                return(self.binsearch(key))
        else:
            return -1

data1 = data([1,3,5,6,7,9,10,14,17,19])
data1.binsearch(10)
```

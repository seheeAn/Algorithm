##### 2.4 matrix multiplication

```python
def matrixmult(n,A,B,C):
    for i in range(n):
        for j in range(n):
            C[i][j] = 0
            for k in range(n):
                C[i][j] = C[i][j]+A[i][k]*B[k][j]
```
